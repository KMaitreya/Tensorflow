```python
#importing required libraries
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
#declaring necessary variables
train_features=[]
train_target=[]

test_features=[]
test_target=[]
```

```python
#Training data generation
#An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial
#The trial had 2100 participants.Hlaf were under the age of 65
#Around 95% of patients 65 or above experienced side effects
#Around 95% of patients under 65 experienced no side effects
#side effects experienced->1, side effects not experienced->0


for i in range(50):
    rand_younger=randint(13, 64)
    train_features.append(rand_younger)
    train_target.append(1)

    rand_older=randint(65, 100)
    train_features.append(rand_older)
    train_target.append(0)

for i in range(1000):
    rand_younger=randint(13, 64)
    train_features.append(rand_younger)
    train_target.append(0)

    rand_older=randint(65, 100)
    train_features.append(rand_older)
    train_target.append(1)
```

```python
#shuffling our data
train_target=np.array(train_target)
train_features=np.array(train_features)
train_target, train_features=shuffle(train_target, train_features)
```

```python
#rescaling the features between 0 and 1
scaler=MinMaxScaler(feature_range=(0, 1))
scaled_train_features=scaler.fit_transform(train_features.reshape(-1, 1))
```

```python
for i in scaled_train_features:
    print(i)
```

```python
#creating the model
model=Sequential()
model.add(Dense(units=16, activation='relu', input_shape=(1,)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
#compiling the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
#training the model
model.fit(scaled_train_features, train_target, epochs=30, batch_size=10, verbose=1)
```

```python
#testing data generation

for i in range(10):
    rand_younger=randint(13, 64)
    test_features.append(rand_younger)
    test_target.append(1)

    rand_older=randint(65, 100)
    test_features.append(rand_older)
    test_target.append(0)

for i in range(200):
    rand_younger=randint(13, 64)
    test_features.append(rand_younger)
    test_target.append(0)

    rand_older=randint(65, 100)
    test_features.append(rand_older)
    test_target.append(1)
```

```python
#shuffling our data
test_features=np.array(test_features)
test_target=np.array(test_target)
test_features, test_target=shuffle(test_features, test_target)
```

```python
#scaling our features between 0 and 1
scaled_test_features=scaler.fit_transform(test_features.reshape(-1,1))
```

```python
#predictions
predictions=model.predict(scaled_test_features)
```

```python
for i in predictions:
    print(i)
```

```python
#converting predictions into 0's and 1's
rounded_predictions=np.argmax(predictions, axis=-1)
print(rounded_predictions)
```

```python
#creating the confusion matrix
conf_matrix=confusion_matrix(test_target, rounded_predictions)
print(conf_matrix)
```

```python
#accuracy
accuracy=accuracy_score(test_target, rounded_predictions)
print(accuracy*100)
```
